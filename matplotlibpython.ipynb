```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
y=2*x
```

```python
x
```

```python
y
```

```python
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=2)
plt.show()
```

```python
y1=2*x
y2=3*x
```

```python
y1
```

```python
y2
```

```python
plt.plot(x,y1,color='r',linestyle=":",linewidth=3)
plt.plot(x,y2,color='g',linestyle="-",linewidth=2)
plt.title("Red vs Green")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x,y1,color='r',linestyle=":",linewidth=3)
plt.plot(x,y2,color='g',linestyle="-.",linewidth=2)
plt.title("Red vs Green")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='r',linestyle=":",linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='g',linestyle=":",linewidth=2)

plt.show()
```

```python
y3=4*x
```

```python
y3
```

```python
plt.subplot(3,1,1)
plt.plot(x,y1,color='r',linestyle=":",linewidth=2)

plt.subplot(3,1,2)
plt.plot(x,y2,color='g',linestyle=":",linewidth=2)

plt.subplot(3,1,3)
plt.plot(x,y3,color='y',linestyle=":",linewidth=2)

plt.show()
```

```python
plt.subplot(3,3,1)
plt.plot(x,y1,color='r',linestyle=":",linewidth=2)

plt.subplot(3,3,2)
plt.plot(x,y2,color='g',linestyle=":",linewidth=2)

plt.subplot(3,3,3)
plt.plot(x,y3,color='y',linestyle=":",linewidth=2)

plt.show()
```

```python
#bar graph
```

```python
students={"Iesha": 70,"Sakshi":80,"Trushna":90}
name=list(students.keys())
marks=list(students.values())
```

```python
plt.bar(name,marks,color='r')
plt.title("Marks Distrubution")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(name,marks,color='r')
plt.title("Marks Distrubution")
plt.xlabel("Marks")
plt.ylabel("Names")
plt.grid(True)
plt.show()
```

```python
#scatter
```

```python
y5=np.array([0,6,3,1,6,3,9,1,4,3])
plt.scatter(x,y5,color='r')
plt.show()
```

```python
y5=np.array([0,6,3,1,6,3,9,1,4,3])
plt.scatter(x,y5,marker="*",color='r',s=100)
plt.show()
```

```python
y6=np.array([1,6,3,2,0,8,2,5,4,8])
```

```python
plt.scatter(x,y5,marker="*",color='r',s=200)
plt.scatter(x,y6,marker=".",color='b',s=200)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y5,marker="*",color='r',s=200)
plt.subplot(1,2,2)
plt.scatter(x,y6,marker=".",color='b',s=200)
plt.show()
```

```python
#pie chart
```

```python
fruit=['Apple','Orange','Mango','Guava']
quantity=[67,34,100,29]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['green','red','gray','black'])
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=2,autopct='%0.1f%%')
plt.pie([1],colors='white',radius=1)
plt.show()
```
